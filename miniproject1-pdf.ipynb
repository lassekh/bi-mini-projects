## Import libraries

```python
# Import from the library PyPDF2 to read PDF
from PyPDF2 import PdfReader
# Library to read tabula data and convert to pandas dataframe
import tabula
import pandas as pd
from collections import Counter
```

## Read from a PDF-file

```python
# Define a path to the pdf
path = 'novo-nordisk-annual-report-2023.pdf'
# Read from that path
reader = PdfReader(path)
```

```python
# Extract text from a page, by giving the page index number
page9 = reader.pages[8]
page9_text = page9.extract_text()

# Extract text from all pages
all_pages = reader.pages
for page in all_pages:
    text = text+page.extract_text()
```

## Count the number of times each word appear in the PDF
..and how put the data into a pandas dataframe

```python
# Count words
text_lower = text.lower()

# Split the text into words
words = text_lower.split()

# Count occurrences of each word
count = Counter(words)

# Convert the word counts to a pandas DataFrame
word_count_df = pd.DataFrame(count.items(), columns=['Word', 'Count'])

# Sort the dataframe by count in desc order
word_count_df = word_count_df.sort_values(by='Count', ascending=False).reset_index(drop=True)
```
